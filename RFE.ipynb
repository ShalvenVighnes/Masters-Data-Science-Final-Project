```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#df is is the original file
#df2 is the data cleaning is done and will be used for the model building
df = pd.read_excel('Final.xlsx')
# Drop constant features
constant_features = df.columns[df.var() == 0]

df.drop(columns=constant_features, inplace=True)
columns_to_drop = ['Patient_Registration_Number', 'Residential_Postcode', 'latitude','longitude']

# Drop the specified columns
df = df.drop(columns=columns_to_drop)

# Display the DataFrame after dropping columns
print(df.head())
```

```python
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Assuming 'target_column' is your target variable column name
target_column = 'AACVPR_Risk_Category'
X = df.drop(columns=[target_column])
y = df[target_column]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=22)

# Initialize the classifier (multi_class='multinomial' for multi-class)= 37 feature.74.28 accuracy
classifier = LogisticRegression(multi_class='multinomial', solver='lbfgs')

best_accuracy = 0
best_n_features = 0
best_selected_features = []

# Iterate through the number of features to select
for n_features in range(1, len(X_train.columns) + 1):
    # Initialize RFE with the classifier and current number of features to select
    rfe = RFE(classifier, n_features_to_select=n_features)

    # Fit RFE
    rfe.fit(X_train, y_train)

    # Get the selected features
    selected_features = X_train.columns[rfe.support_]

    # Train a model with the selected features
    classifier.fit(X_train[selected_features], y_train)

    # Make predictions on the test set
    y_pred = classifier.predict(X_test[selected_features])

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)

    # Update best accuracy and features if the current iteration is better
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_n_features = n_features
        best_selected_features = selected_features

    print(f"Number of Features: {n_features}, Accuracy: {accuracy}")

# Display the best results
print("\nBest Results:")
print(f"Number of Features: {best_n_features}")
print("Selected Features:", best_selected_features)
print("Best Accuracy:", best_accuracy)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Assuming 'target_column' is your target variable column name
target_column = 'AACVPR_Risk_Category'
X = df.drop(columns=[target_column])
y = df[target_column]


# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=22)

# List of selected features
selected_features = ['Patient_occupation', 'Prescribed_Sessions', 'HDL', 'HbA1c',
                     'Post_Tobacco', 'Pre_Weight', 'Height', 'Pre_Peak_Heart_Rate',
                     'Post_Exercise_Stress_Test', 'Post_Peak_METs',
                     'Exercise_frequency_sessions_week', 'Exercise_intensity', 'Post_RTW',
                     'MACCE', 'Age', 'Gender', 'Pre_Left_Ventricle_EF',
                     'Admission_Diagnosis_NSTEMI', 'Admission_Diagnosis_STEMI',
                     'Admission_Diagnosis_Valve replacement(s)',
                     'Risk_Factors_High Lipid Profile', 'Risk_Factors_Hypertension',
                     'Risk_Factors_Unknown', 'Past_CV_None of the above', 'Past_CV_Prev MI',
                     'Pre_BP_systolic', 'Pre_BP_diastolic', 'Pre_Medication_Aspirin',
                     'Pre_Medication_Statin', 'Post_Medication_ACEI/ARB',
                     'Post_Medication_Aspirin', 'CR_Medication_Aspirin',
                     'Duration_Between_Ward_Enrollment', 'Pre_BMI', 'Pre_METs',
                     'Pre_METs_range', 'Post_Peak_METs_range']

# Initialize the Logistic Regression classifier
classifier = LogisticRegression()

# Define hyperparameter grid for grid search
param_grid = {
    'C': [0.0001,0.001, 0.01, 0.1, 1, 10, 100,1000],
    'penalty': ['l1', 'l2']
}

# Initialize GridSearchCV
grid_search = GridSearchCV(classifier, param_grid, cv=10, scoring='accuracy')

# Perform grid search on the selected features
grid_search.fit(X_train[selected_features], y_train)

# Get the best parameters and model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Make predictions on the test set
y_pred = best_model.predict(X_test[selected_features])

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Best Parameters: {best_params}")
print(f"Model Accuracy on Test Set: {accuracy}")

# Additional evaluation metrics
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
